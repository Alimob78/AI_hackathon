````python
from langchain.output_parsers import StructuredOutputParser, ResponseSchema

response_schemas = [
    ResponseSchema(name="response", description="response to the given prompt"),
    ResponseSchema(name="translation", description="translated response"),
    ResponseSchema(name="summary", description="brief summary of the conversation so far (in English)")
]
output_parser = StructuredOutputParser.from_response_schemas(response_schemas)
format_instructions = output_parser.get_format_instructions()
````

````python
from langchain import PromptTemplate

template = """
You are the founder of a top tier tech company and coaching the given candidate based on what their goal is.

A mentor would answer their mentees questions as much as geniunly as they can based on the given context,
and summary of previous chat history.
If you cannot answer a question try to recommend a book or other famous person in tech.
First generate a response to the question in English in less that about 50 words,
then translate the response to {language},
and then generate an overall summary of the conversation so far along with your last response in English.
Your response should be in the following format:
1. response to the given prompt,
2. Translation in {language}, and
3. A brief summary of the previous chat history and current conversation (in English).

Previous chat history: ```{chat_history}'''
Question: ```{query}'''
Context: ```{context}'''

{format_instructions}
"""

# Format the output as a JSON with the following keys:
# Response
# Translation
# Summary

prompt_template = PromptTemplate(
    input_variables=["language", "chat_history", "query", "context"],
    template=template,
    partial_variables={"format_instructions": format_instructions}
)
````

````python
first_prompt = prompt_template.format(
    language="Farsi",
    chat_history="",
    query=query,
    context=content
)

print(first_prompt)
````

````python
import pandas as pd
import random
from faker import Faker

# Generate random education start and end dates
def generate_education_dates():
    start_year = random.randint(2000, 2020)
    end_year = random.randint(start_year, 2022)
    return start_year, end_year

# Generate random experience start and end dates
def generate_experience_dates():
    start_year = random.randint(2000, 2020)
    end_year = random.randint(start_year, 2022)
    return start_year, end_year

# Generate a summary with over 100 words
def generate_summary():
    fake = Faker()
    summary = fake.paragraphs(nb=5)
    while len(summary) < 70:
        summary = fake.paragraphs(nb=5)
    return ' '.join(summary)

# Dummy data generation
data = {
    'Name': [],
    'Headline': [],
    'Location': [],
    'Connections': [],
    'Summary': [],
    'Experience': [],
    'Experience Start Date': [],
    'Experience End Date': [],
    'Experience Role': [],
    'Experience Responsibility': [],
    'Education': [],
    'Education Start Date': [],
    'Education End Date': []
}

computer_science_fields = [
    'Software Engineering',
    'Data Science',
    'Artificial Intelligence',
    'Cybersecurity',
    'Web Development',
    'Database Administration',
    'Network Engineering'
]

for i in range(10):
    name = f'Person {i+1}'
    headline = random.choice(computer_science_fields)
    location = 'Unknown'
    connections = random.randint(100, 1000)
    summary = generate_summary()

    experience = f'{headline} Engineer at ABC Company'
    exp_start, exp_end = generate_experience_dates()
    experience_role = f'{headline} Engineer'
    experience_responsibility = 'Developed and implemented innovative solutions.'

    education = f'Bachelor of {headline} at University of XYZ'
    edu_start, edu_end = generate_education_dates()

    data['Name'].append(name)
    data['Headline'].append(headline)
    data['Location'].append(location)
    data['Connections'].append(connections)
    data['Summary'].append(summary)
    data['Experience'].append(experience)
    data['Experience Start Date'].append(exp_start)
    data['Experience End Date'].append(exp_end)
    data['Experience Role'].append(experience_role)
    data['Experience Responsibility'].append(experience_responsibility)
    data['Education'].append(education)
    data['Education Start Date'].append(edu_start)
    data['Education End Date'].append(edu_end)

df = pd.DataFrame(data)
df.head()
````
